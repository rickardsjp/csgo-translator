```python
from PIL import Image, ImageOps
import pytesseract
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(3,3))
```

```python
img = cv2.imread('../img/test2.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, interpolation='none', cmap='gray')
```

```python
text_area = gray[800:1000, 170:614]
plt.imshow(text_area, interpolation='none', cmap='gray')
```

## Image Conversion

```python
kernel = np.ones((2,1),np.uint8)
kernel2 = np.ones((6,5),np.uint8)
```

```python
thr3 = cv2.adaptiveThreshold(text_area,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,39,-145)
eroded = cv2.erode(thr3,kernel,iterations = 1)
dilated = cv2.dilate(eroded,kernel2,iterations = 3)
```

```python
Image.fromarray(dilated).show()
extracted = cv2.bitwise_and(text_area,text_area, mask = dilated)
Image.fromarray(extracted).show()
```
